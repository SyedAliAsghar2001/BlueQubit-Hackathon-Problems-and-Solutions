```python
from qiskit import QuantumCircuit

qc = QuantumCircuit.from_qasm_file(r"C:\Users\syeda\OneDrive\Desktop\BlueQubit Hackathon\P3__sharp_peak.qasm")

print(f"Circuit depth: {qc.depth()}")
print(f"Number of gates: {qc.size()}")
print(f"Qubits: {qc.num_qubits}")

qc.measure_all()
```

```python
import bluequbit
bq = bluequbit.init("nyxHwJVZQfA3Ko48lSYdKrPvTjVcVkEK")
result = bq.run(qc, device = "mps.cpu")
```

```python
counts = result.get_counts()
print(counts)
```

```python
peaked_state = max(counts, key=counts.get)
max_count = counts[peaked_state]
probability = (max_count / 2000) * 100  # Convert to percentage

print(f"Peaked state: {peaked_state} | Count: {max_count} | Probability: {probability:.2f}%")
```
